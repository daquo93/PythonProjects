```python
from turtle import Turtle, Screen
import time
import random
from car import Car
screen = Screen()
width = 800
height = 600
screen.setup(width=width, height=height)
screen.bgcolor("white")
screen.title("Cross road game")
screen.tracer(0)

my_turtle = Turtle()
my_turtle.shape("turtle")
my_turtle.penup()
my_turtle.hideturtle()
my_turtle.goto(0,-height/2-20)
my_turtle.showturtle()
my_turtle.setheading(90)
screen.update()
screen.listen()

def move_forward():
    my_turtle.fd(20)

# define a function for moving the turtle backward
def move_backward():
    my_turtle.fd(-20)

screen.onkey(move_forward,"Up")
screen.onkey(move_backward,"Down")




game_is_on = True
my_cars = []

def initialize_cars():
    for i in range(280,-280,-40):
        if random.randint(1,3) % 3 == 0:
            c = Car()
            c.initialize_car(i)
            my_cars.append(c)
    return my_cars
            
        
    

def move_cars():
    global my_cars
    for car in my_cars:
        car.move()
        

def detect_collision():
    
    global my_turtle
    global my_cars
    collision = False
    for car in my_cars:
        if car.distance(my_turtle)<=20:
            collision = True
    return collision
    

my_cars = initialize_cars()
print(len(my_cars))
counter = 0
while game_is_on:
    counter+=1
    screen.update()
    time.sleep(0.1)
    move_cars()
    if detect_collision():
        game_is_on = False
        break
    if counter % 10 == 0:
        my_cars = initialize_cars()
    for car in my_cars:
        if car.xcor()<=0:
            car.delete_car()
        
        
    
    
    
    



        
                    
                    
        
        
    















screen.mainloop()
```
