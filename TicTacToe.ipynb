```python
import numpy as np
grid = "_|_|_\n_|_|_\n | | "
locations = []
locations1 = np.zeros((3, 3))
locations2 = np.zeros((3, 3))
game_is_on = True
def convert_coord_to_int(coord):
    x = int(coord.split(",")[0])
    y = int(coord.split(",")[1])
    return 6*x+y*2

def is_won(loc):
    if np.fliplr(loc).diagonal().sum() == 3 or np.diagonal(loc).sum() ==3 or 3 in loc.sum(axis = 0) or 3 in loc.sum(axis = 1):
        return True
    else:
        return False
    

while game_is_on:
    coord1accepted = False
    coord2accepted = False
    while not coord1accepted:
        c1 = input("Player one, chose a location to place your cross; provide row and column in the format row,col\n\n\n")
        if (int(c1.split(",")[0]) >=0 and int(c1.split(",")[0]) <=2) and (int(c1.split(",")[1]) >=0 and int(c1.split(",")[1] )<=2):
                location = convert_coord_to_int(c1)
                if location not in locations:
                    locations.append(location)
                    locations1[int(c1.split(",")[0]),int(c1.split(",")[1])] = 1
                    coord1accepted = True
                    grid = grid[:location]+"x"+grid[location+1:]
                    print(grid)
    
    if is_won(locations1):
        print("Player 1 won")
        game_is_on = False
        break
                     
    while not coord2accepted:
        c1 = input("Player two, chose a location to place your cross; provide row and column in the format row,col\n\n\n")
        if (int(c1.split(",")[0]) >=0 and int(c1.split(",")[0]) <=2) and (int(c1.split(",")[1]) >=0 and int(c1.split(",")[1]) <=2):
                location = convert_coord_to_int(c1)
                if location not in locations:
                    locations.append(location)
                    locations2[int(c1.split(",")[0]),int(c1.split(",")[1])] = 1
                    coord2accepted = True
                    grid = grid[:location]+"o"+grid[location+1:]
                    print(grid)
                    
    if is_won(locations2):
        print("Player 2 won")
        game_is_on = False
        break  
        
    if len(locations) == 9:
        print("No one won!")
        game_is_on = False
        break
    
    
    
    
    
                    
                    
                    
                
            
    
```
