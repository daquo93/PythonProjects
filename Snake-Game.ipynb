```python
from turtle import Turtle, Screen
import time
import random
screen = Screen()
width = 600
screen.setup(width=width, height=width)
screen.bgcolor("black")
screen.title("My snake game")
screen.tracer(0)

my_turtles = []
treasure = Turtle()
treasure.shape("square")
treasure.color("blue")

def show_treasure():
    
    treasure.penup()
    treasure.hideturtle()
    stay_in = True
    while stay_in:
        treasure.goto(random.randrange(-width/2,width/2,20),random.randrange(-width/2,width/2,20))
        collision = False
        for tur in my_turtles:
            if treasure.distance(tur) == 0:
                collision = True
            if not collision:
                stay_in = False
    treasure.showturtle() 
    
    
def check_collision():
    global my_turtles
    for i in range(len(my_turtles)):
        for j in range(i+1, len(my_turtles)):
            if my_turtles[i].distance(my_turtles[j]) < 10:
                screen.bgcolor("red")
                
                message = turtle.Turtle()
                message.hideturtle()
                message.write("GAME OVER", align="center", font=("Courier", 24, "bold"))
    
    # pause the game for 2 seconds
                screen.update()
                turtle.delay(2000)
    
    # exit the game
                turtle.bye()
        
    



x_pos = 20
for i in range(0, 3):
    t = Turtle()
    t.shape("square")
    t.color("white")
    t.penup()
    t.goto(x_pos, 0)
    x_pos -= 20
    my_turtles.append(t)
    screen.update()
    
    
def add_block(x_last,y_last):
    global my_turtles
    global treasure
    if my_turtles[0].distance(treasure) < 10:
            t = Turtle()
            t.shape("square")
            t.color("white")
            t.penup()
            t.goto(x_last,y_last)
            my_turtles.append(t)
            screen.update()
            show_treasure()
    
    

stop_for = False
stop_back = False
stop_up = False
stop_down = False

def stop_backward():
    global stop_back
    stop_back = True
    return stop_back

def stop_forward():
    global stop_for
    stop_for = True
    return stop_for

def stop_upward():
    global stop_up
    stop_up = True
    return stop_up

def stop_downward():
    global stop_down
    stop_down = True
    return stop_down

screen.onkey(stop_backward, 'w') 
screen.onkey(stop_backward, 's') 
 

screen.onkey(stop_forward, 'w') 
screen.onkey(stop_forward, 's') 

screen.onkey(stop_upward, 'd') 
screen.onkey(stop_upward, 'a')


screen.onkey(stop_downward, 'd') 
screen.onkey(stop_downward, 'a')


game_started = False

def move_forward():
    global width
    global stop_for
    global my_turtles
    
    if my_turtles[0].heading() == 180:
            return
    while not stop_for:
        x_last = my_turtles[-1].xcor()
        y_last = my_turtles[-1].ycor()
        
        x_initial = my_turtles[0].xcor()
        y_initial = my_turtles[0].ycor()
        
        my_turtles[0].setheading(0)
        my_turtles[0].fd(20)
        check_collision()
        
        if my_turtles[0].distance(treasure) < 10:
            t = Turtle()
            t.shape("square")
            t.color("white")
            t.penup()
            t.goto(x_last,y_last)
            my_turtles.append(t)
            screen.update()
            show_treasure()
            
        
        if(my_turtles[0].xcor()> width/2):
            my_turtles[0].goto(-width/2,my_turtles[0].ycor())
            
        
        for i in range(1,len(my_turtles)):
            x_temp = my_turtles[i].xcor()
            y_temp = my_turtles[i].ycor()
            
            
            my_turtles[i].goto(x_initial,y_initial)
            x_initial = x_temp
            y_initial = y_temp
        
        screen.update()
        time.sleep(0.2)
        
    stop_for = False
    
    

    
def move_backward():
    global width
    global stop_back
    global my_turtles
    
    if my_turtles[0].heading() == 0:
            return
    while not stop_back:
        x_last = my_turtles[-1].xcor()
        y_last = my_turtles[-1].ycor()
        
        x_initial = my_turtles[0].xcor()
        y_initial = my_turtles[0].ycor()
        
        my_turtles[0].setheading(180)
        my_turtles[0].fd(20)
        check_collision()
        
        add_block(x_last,y_last)
        
        if(my_turtles[0].xcor()< -width/2):
            my_turtles[0].goto(width/2,my_turtles[0].ycor())
            
        
        for i in range(1,len(my_turtles)):
            x_temp = my_turtles[i].xcor()
            y_temp = my_turtles[i].ycor()
            
            
            my_turtles[i].goto(x_initial,y_initial)
            x_initial = x_temp
            y_initial = y_temp
        
        screen.update()
        time.sleep(0.2)
        
    stop_back = False
    
    
    
    
    
    
    
        
        
        
 
    
def move_downward():
    
    global width
    global stop_down
    global my_turtles
    
    if my_turtles[0].heading() == 90:
            return
    while not stop_down:
        x_last = my_turtles[-1].xcor()
        y_last = my_turtles[-1].ycor()
        x_initial = my_turtles[0].xcor()
        y_initial = my_turtles[0].ycor()
        
        my_turtles[0].setheading(270)
        my_turtles[0].fd(20)
        check_collision()
        
        add_block(x_last,y_last)
        
        if(my_turtles[0].ycor()< -width/2):
            my_turtles[0].goto(my_turtles[0].xcor(),width/2)
            
        
        for i in range(1,len(my_turtles)):
            x_temp = my_turtles[i].xcor()
            y_temp = my_turtles[i].ycor()
            
            
            my_turtles[i].goto(x_initial,y_initial)
            x_initial = x_temp
            y_initial = y_temp
        
        screen.update()
        time.sleep(0.2)
        
    stop_down = False
        
def move_upward():
    
    global width
    global stop_up
    global my_turtles
    
    if my_turtles[0].heading() == 270:
            return
    while not stop_down:
        x_last = my_turtles[-1].xcor()
        y_last = my_turtles[-1].ycor()
        x_initial = my_turtles[0].xcor()
        y_initial = my_turtles[0].ycor()
        
        my_turtles[0].setheading(90)
        my_turtles[0].fd(20)
        check_collision()
        add_block(x_last,y_last)
        if(my_turtles[0].ycor()> width/2):
            my_turtles[0].goto(my_turtles[0].xcor(),-width/2)
            
        
        for i in range(1,len(my_turtles)):
            x_temp = my_turtles[i].xcor()
            y_temp = my_turtles[i].ycor()
            
            
            my_turtles[i].goto(x_initial,y_initial)
            x_initial = x_temp
            y_initial = y_temp
        
        screen.update()
        time.sleep(0.2)
        
    stop_up = False
        
    
   
  
    
    
show_treasure()


screen.onkey(move_forward, 'd')
screen.onkey(move_backward, 'a')
screen.onkey(move_downward, 's')
screen.onkey(move_upward, 'w')

screen.listen()
screen.mainloop()
```
